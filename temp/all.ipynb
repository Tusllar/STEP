```python
import h5py
import pandas as pd
import os

# Đường dẫn đến file .h5
input_file = r"D:\PBL4\STEP\Data EWalk\Labels.h5"
output_folder = "output_csv2"

# Tạo thư mục nếu chưa có
os.makedirs(output_folder, exist_ok=True)

# Mở file .h5
with h5py.File(input_file, 'r') as f:
    # Lấy danh sách keys (các nhóm dữ liệu)
    keys = list(f.keys())
    print(f"Tổng số nhóm dữ liệu: {len(keys)}")

    for key in keys:
        data = f[key][()]  # lấy toàn bộ dữ liệu từ nhóm
        try:
            df = pd.DataFrame(data)
            output_path = os.path.join(output_folder, key)
            df.to_csv(output_path, index=False)
            print(f"Đã lưu: {output_path}")
        except Exception as e:
            print(f"Lỗi ở {key}: {e}")
```
